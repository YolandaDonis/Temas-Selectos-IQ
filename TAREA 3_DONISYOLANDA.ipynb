EJERCICIO 1

```python
#Cálculo de la presión de un gas ideal.

#Datos
n = 2.0 #mol
R = 0.0821 #L*atm/(mol*K)
T = 298 #K
V = 10 #L

#Ecuación de gas ideal
P = n*R*T/V

print(f"Presión del gas en atm: {P:.2f}")
```

EJERCICIO 2

```python
#Cálculo de la constante de velocidad k con la ecuación de Arrhenius

#Datos
A = 50000000 #1/s
Ea = 75000 #J/mol
R = 8.341 #J/(mol*K)
T = 350 #K

import math

#Ecuación de Arrhenius
k = math.exp(-Ea/(R*T))

print(f"El valor de la constante de velocidad es:{k}")
```

EJERCICIO 3

```python
#Cálculo de la presión de vapor empleando la ecuación de Antoniete

#Datos
A = 8.07131
B = 1730.63
C = 233.426
T = 80 #°C

#Ecuación de Antoniete
P = 10**(A-(B/(C+T)))

print(f"Presión de vapor en mmHg: {P:.2f}" )
```

EJERCICIO 4

```python
#Cálculo del número de Reynolds

#Datos
rho = 1000 #kg/m3
v = 1.2 #m/s
D = 0.05 #m
vis = 0.001 # Pa*s

#Ecuación
Re = rho*v*D/vis

print(f"El valor del Reynolds es: {Re:.2f}" )
```

EJERCICIO 5

```python
#Determinar el tipo de flujo de un fluido

rho = float(input("Densidad del fluido en kg/m3 "))
v = float(input("Velocidad promedio del fluido en m/s "))
D = float(input("Diámetro interno de la tubería en m "))
vis = float(input("Viscosidad dinámica en Pa*s "))

#ecuación
Re = rho*v*D/vis

print(f"El valor del Reynolds es: {Re:.2f}" )

#Determinar el régimen de flujo
if Re <= 2100:
    print("flujo laminar")
elif Re > 2100 and Re < 4000:
    print("estado transitorio")
else:
    print ("flujo turbulento")
```
