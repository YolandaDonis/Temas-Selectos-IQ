Cdigo que te haga un IF aceptando las unidades que ponga el usuario. Que cambie de celsius o farhenheit a K

```python
def T_K(T,unidad):
  if unidad == "K":
    return T
  elif unidad =="C":
    return T+273.15
  elif unidad =="F":
    return (T-32)*5/9+273.15
  elif unidad == "R":
    return T*5/9
  else:
    print("Data no vlida")
```

```python
T_K(32,"K")
```

```python
def P_Mpa(P,unidad):
  if unidad == "Mpa":
    return P
  elif unidad =="bar":
    return P*0.1
  elif unidad =="mmHg":
    return P*0.000133322
  elif unidad == "Psi":
    return P*0.00689476
  else:
    print("Data no vlida")
```

```python
P_Mpa(1,"mmHg")
```

```python
def definir_estado(T_Kelvin, P_mpa):
  Tc = 647.1
  Pc = 22.06

  if T_Kelvin<Tc and P_mpa<Pc:
    return "Lquido o vapor"
  elif T_Kelvin>Tc and P_mpa>Pc:
    return "Supercrtica"
  elif T_Kelvin<Tc and P_mpa>Pc:
    return "Lquido comprimido"
  else:
    return "Cerca del punto crtico"
```

```python
capitales = ['#Ciudad Victo*ria', 'cuerna#vaca', 'Morelia!', 'xalapa', 'Tepic', 'hermos*illo',
 'toluca?', 'San? Luis Potos', '*Mexicali', 'Villahermosa', 'Tuxtla Gutirrez#',
 'Colima', 'la Paz#', 'Guadalajara', 'C*ampeche!', 'Durango', 'Zacatecas',
 '!Chetumal', 'Ciudad# de Mxico', 'chilpancingo*', 'Chihua#hua', '?Aguascalientes',
 'Oaxaca de Jurez*', '?Puebla', 'Quer?taro', 'Mrida', 'guanajuato',
 'pachu*ca', '#Saltillo#', 'Culiacn!', 'Tlaxcala', 'monterrey']
```

```python
import re
def limpiar_cadenas(strings):
  result = []
  for value in strings:
    value=value.strip() #suprime blancos y otro caracter especificado
    value=re.sub("[#]","",value) #remplaza todas las ocurrencias de un poatrn
    value=value.title() #convierte la primera letra de cada palabra de una cadena
    result.append(value)
  return result
```

```python
limpiar_cadenas(capitales)
```

```python
capitales_limpias= limpiar_cadenas(capitales)
capitales_limpias.sort() #para ordenar alfabticamente

capitales_limpias
```
