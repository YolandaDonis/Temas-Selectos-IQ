```python
ruta_poblacion ="https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/poblacion.csv"
poblacion = pd.read_csv(ruta_poblacion)
poblacion.head()
```

```python
poblacion['1990'] = poblacion['H_1990'] + poblacion['M_1990']
poblacion['2000'] = poblacion['H_2000'] + poblacion['M_2000']
poblacion['2010'] = poblacion['H_2010'] + poblacion['M_2010']
poblacion['2020'] = poblacion['H_2020'] + poblacion['M_2020']
```

```python
poblacion = poblacion.drop(['H_1990','M_1990','H_2000',
                            'M_2000','H_2010','M_2010','H_2020',
                            'M_2020','Regiones'],axis=1)      #el drop es para eliminar las columnas que no m interesa usar
poblacion.head()
```

```python
poblacion.set_index("Estado") #para añadir un índice
```

```python
F =["1990","2000","2010","2020"]
C=[poblacion['1990'].sum(),poblacion['2000'].sum(),
   poblacion['2010'].sum(),poblacion['2020'].sum()]
```

```python
pob = pd.DataFrame({'Poblacion' : C}, index=F) #para que lo reconozca como X y Y
```

```python
plt.bar(pob.index, pob['Poblacion'], color = "red", alpha=0.8) #alpha es para la intensidad del color
plt.title('Poblacion por año')
plt.ylabel('Poblacion (100,000,000)')
plt.xlabel('Año')
plt.show()
```

```python
ruta= "https://raw.githubusercontent.com/marinlopabi/Temas-Selectos-PQ-AD-Python/refs/heads/main/Datos/CEMEXCPO.csv"
df = pd.read_csv(ruta) #lee el archivo como un csv
df.head()
```

```python
df.info()
```

```python
import pandas as pd
```

```python
#para reconozca el dato como serie de tiempo(la columna de datos)

df["Date"] = pd.to_datetime(df["Date"])
```

```python
df.info()
```

```python
#para que cambie el índice que pone por default y lo sustituya por la columna que indico
df = df.set_index("Date")
df
```

Graficar

```python
plt.figure(figsize=(16,8), dpi=150)
plt.plot(df.index,df["Close"],lw=2.5,color="black",label="Precio de cierre") #línea principal de cierre #lw para el grosor de la linea #label para poner leyenda
#para etiquetas y títulos
plt.xlabel("Fecha", fontsize=18, labelpad=10) #fontisize para el tamaño de las letras #labelpad para la separación
plt.ylabel("Precio($)", fontsize=18, labelpad=10)
plt.title('Precio de Cierre - CEMEX', fontsize=22, fontweight='bold', pad=15) #weigjt para letras en negritas

#configuración del gráfico
plt.grid(alpha=0.3) #para transparencia
plt.tick_params(labelsize=14) #tamaño de ejes
plt.legend(fontsize=14) #tamaño de leyenda
plt.tight_layout()
plt.show()
```

```python
# Estilo general más profesional
plt.style.use('seaborn-v0_8-whitegrid')  # mejora la estética del gráfico

plt.figure(figsize=(16, 8), dpi=150)

# ==========================
# LÍNEAS PRINCIPALES
# ==========================
plt.plot(df.index, df["Close"], lw=2.5, color="black", label="Precio de cierre")  # línea principal

# Promedio móvil (ventana de 20 días) para ver la tendencia
df["MA20"] = df["Close"].rolling(window=20).mean()
plt.plot(df.index, df["MA20"], color="blue", lw=2, linestyle="--", label="Media móvil (20 días)")

# ==========================
# PUNTOS CLAVE (MÍNIMO y MÁXIMO)
# ==========================
min_idx = df["Close"].idxmin()  # índice del valor mínimo
max_idx = df["Close"].idxmax()  # índice del valor máximo

# Marcamos los puntos en el gráfico
plt.scatter(min_idx, df["Close"].min(), color="red", s=80, zorder=5, label="Valor mínimo")
plt.scatter(max_idx, df["Close"].max(), color="green", s=80, zorder=5, label="Valor máximo")

# Anotaciones con valores
plt.annotate(f"Mín: ${df['Close'].min():.2f}",
             xy=(min_idx, df["Close"].min()),
             xytext=(-80, -30), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="red"), fontsize=12, color="red")

plt.annotate(f"Máx: ${df['Close'].max():.2f}",
             xy=(max_idx, df["Close"].max()),
             xytext=(40, 30), textcoords="offset points",
             arrowprops=dict(arrowstyle="->", color="green"), fontsize=12, color="green")

# ==========================
# SOMBREADO ENTRE PRECIO ALTO Y BAJO (SI EXISTEN)
# ==========================
if "High" in df.columns and "Low" in df.columns:
    plt.fill_between(df.index, df["Low"], df["High"], color="gray", alpha=0.15, label="Rango diario")

# ==========================
# CONFIGURACIÓN DE ETIQUETAS Y TÍTULOS
# ==========================
plt.xlabel("Fecha", fontsize=18, labelpad=10)
plt.ylabel("Precio ($)", fontsize=18, labelpad=10)
plt.title("Evolución del precio de cierre - CEMEX", fontsize=22, fontweight="bold", pad=15)

# ==========================
# FORMATOS DE EJES Y LEYENDAS
# ==========================
plt.xticks(rotation=45)  # rotamos fechas para legibilidad
plt.grid(alpha=0.3)
plt.tick_params(labelsize=14)
plt.legend(fontsize=13, loc="best", frameon=True)  # leyenda mejor posicionada

# Añadimos línea horizontal en el promedio general
plt.axhline(df["Close"].mean(), color="orange", lw=1.5, linestyle=":", label="Promedio general")

plt.tight_layout()  # evita solapamiento de etiquetas
plt.show()
```
