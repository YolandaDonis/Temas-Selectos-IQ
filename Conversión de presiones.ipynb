```python
def P_Mpa(P,unidad):
  if unidad == "Mpa":
    return P
  elif unidad =="bar":
    return P*0.1
  elif unidad =="mmHg":
    return P*0.000133322
  elif unidad == "Psi":
    return P*0.00689476
  elif unidad=="atm":
    return P*0.101325
  else:
    print("Data no válida")
```

```python
P_Mpa(1,"atm")
```
