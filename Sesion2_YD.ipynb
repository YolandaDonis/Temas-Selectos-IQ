Para resolver ecuaciones la librería de math es una herramienta

```python
a = -1
b = 2
c = 3
import math #para resolver ecuaciones con raiz cuadrada se importa ese modulo math
x1 = ((-b)+math.sqrt(b**2-4*a*c))/2*a  #ecuacion del discriminante
x2 = ((-b)-math.sqrt(b**2-4*a*c))/2*a
```

```python
print(x1, x2)
```

```python
n = int(input('ingresa un número: ')) #input permite al usuario ingresar datos y el int ayuda a que lo reconozca como número y no como texto
if n>0:
  print('el número es positivo')
elif n<0:
  print('el número es negativo')
else:
  print('el número es cero')
```

```python
#declarando las variables
a = 1
b = 4
c = 5
#escribiendo la ecuación
D = b**2-4*a*c
if D>0:
  print('son raíces reales distintas')
elif D<0:
  print('son raíces complejas conjugadas')
else:# este no lleva condicion ya que si no se cumple ni una ni otra ya se cumple lo que queda
  print('D=0 son raíces doble')
```
