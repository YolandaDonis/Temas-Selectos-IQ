Insertar la ecuación de gas ideal para realizar el cálculo (dando valores a las variables)


```python
#definiendo variables
n = 2.0
R = 0.081
T = 298
V = 10

#ecuacion-gas-ideal
P = n*R*T/V

print(P)
```

```python
import math #la librería de math puede ayudar con ciertas operaciones matemáticas más complejas
#constantes
A = 8.07131
B = 1730.63
C = 233.426
T = 80 #°c

#ecuación
P = 10**(A-(B/(C+T)))
print(f"Presión de vapor en °C: {P:.2f}" )
```

```python
#cálculo del número de Reynolds

import math
rho = 1000 #kg/m3
v = 1.2 #m/s
D = 0.05 #m
vis = 0.001 # Pa*s

#ecuación
Re = rho*v*D/vis

print(f"El valor del Reynolds es: {Re:.2f}" )

#para saber en que régimen está
if Re <= 2100:
    print("flujo laminar")
elif Re > 2100 and Re < 4000:
    print("estado transitorio")
else:
    print ("flujo turbulento")
```
